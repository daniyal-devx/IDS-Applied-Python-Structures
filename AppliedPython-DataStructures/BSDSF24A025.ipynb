## Problem 3 – Student Performance Report

```python
student_records = {
    "Ahmed Raza": {"Math": 92, "Physics": 85, "Chemistry": 87},
    "Bilal Khan": {"Math": 88, "Physics": 90, "Chemistry": 82},
    "Saad Malik": {"Math": 85, "Physics": 80, "Chemistry": 85},
    "Zain Abid": {"Math": 90, "Physics": 75, "Chemistry": 69},
}

student_averages = {}
for student, subjects in student_records.items():
    total_marks = sum(subjects.values())
    average = total_marks / len(subjects)
    student_averages[student] = average

sorted_results = sorted(student_averages.items(), key=lambda x: x[1], reverse=True)

print("Top Students:")
for student, avg in sorted_results[:3]:
    print(student, ":", avg)

print("\nSorted Results:")
print(sorted_results)
```

## Problem No 4 - Email Data Processor

```python
email_list = [
    "ahmed.raza@hotmail.com",
    "fatima.khan@yahoo.com",
    "omar.malik@gmail.com",
    "zara.ahmed@outlook.com",
    "invalid-email",
    "admin@pucit.edu.pk",
    "usman123@gmail.com"
]

valid_emails = []
invalid_emails = []
domain_stats = {}

for email_address in email_list:
    if '@' in email_address and '.' in email_address.split('@')[1]:
        valid_emails.append(email_address)
        domain = email_address.split('@')[1]
        if domain in domain_stats:
            domain_stats[domain] += 1
        else:
            domain_stats[domain] = 1
    else:
        invalid_emails.append(email_address)

print("Valid emails:", valid_emails)
print("Invalid emails:", invalid_emails)
print("Domain distribution:", domain_stats)
```

## Problem 6 – Restaurant Billing System

```python
menu = [{"name": "Biryani", "price": 450},
        {"name": "Karahi", "price": 1200},
        {"name": "Daal", "price": 250},
        {"name": "Raita", "price": 100}
        ]
        
order = [("Biryani", 3),
("Karahi", 1),
("Raita", 5)
]

gross_total = 0

for item, quantity in order:
    for menu_items in menu:
        if menu_items["name"] == item:
            gross_total += menu_items["price"] * quantity
            
service_quality = 8

if service_quality >= 8:
    tip = 0.2 * gross_total
elif service_quality >= 5:
    tip = 0.15 * gross_total
else: 
    tip = 0.10 * gross_total

tax = 0.05 * gross_total
net_total = gross_total + tax + tip

print("Your total bill is: ", net_total)
```

## Problem 7 – Transport Schedule Tracker

```python
stops = [
    {"name": "Bahria", "time": 0, "distance": 0},
    {"name": "Defence", "time": 10, "distance": 8},
    {"name": "Johar", "time": 5, "distance": 15},
    {"name": "Model", "time": 5, "distance": 22},
    {"name": "Airport", "time": 20, "distance": 35}
]

num_stops = len(stops)

Total_distance = stops[-1]["distance"] - stops[0]["distance"]
total_time = 0
for stop in stops:
    total_time += stop["time"]
    
shortest_distance = 1000
for i in range(1, num_stops):
    segment_distance = stops[i]["distance"] - stops[i-1]["distance"]
    if segment_distance < shortest_distance:
        shortest_distance = segment_distance

print("Total no of stops: ", num_stops)
print("Total Time Taken: ", total_time)
print("Shortest distance: ", shortest_distance)
```

## Problem 9 – Sports Tournament Statistics (6 marks)

```python
matches = [
    {"team1": "Lahore Qalandars", "team2": "Karachi Kings", "score1": 3, "score2": 2},
    {"team1": "Islamabad United", "team2": "Peshawar Zalmi", "score1": 1, "score2": 1},
    {"team1": "Karachi Kings", "team2": "Islamabad United", "score1": 0, "score2": 4}
]

teams = []
score = []
for match in matches:
    if match["team1"] in teams:
        index = teams.index(match["team1"])
        score[index] += match["score1"]
    else:
        teams.append(match["team1"])
        score.append(match["score1"])
        
    if match["team2"] in teams:
        index = teams.index(match["team2"])
        score[index] += match["score2"]
    else:
        teams.append(match["team2"])
        score.append(match["score2"])
        
max_score = max(score)
winner_index = score.index(max_score)
winner = teams[winner_index]

print("Teams:", teams)
print("Scores:", score)
print("Winning Team:", winner)
```

## Problem 14 – Hotel Booking Management

```python
rooms = {
    101: {"available": False, "cost_per_night": 5000, "nights": 3},
    102: {"available": True, "cost_per_night": 4000, "nights": 0},
    103: {"available": False, "cost_per_night": 4500, "nights": 2},
    104: {"available": True, "cost_per_night": 3500, "nights": 0}
}

total_cost = 0
occupied = []
available = []

for room, info in rooms.items():
    if info["available"]:
        available.append(room)
    else:
        occupied.append(room)
        total_cost += info["cost_per_night"] * info["nights"]

print("Occupied Rooms:", occupied)
print("Available Rooms:", available)
print("Total Stay Cost of Occupied Rooms:", total_cost)
```

## Problem 15 – Expense & Budget Tracker

```python
expenses = [
    {"category": "Food", "amount": 1500},
    {"category": "Transport", "amount": 800},
    {"category": "Entertainment", "amount": 1200},
    {"category": "Food", "amount": 1000},
    {"category": "Bills", "amount": 2000}
]

budget = 6000
category_totals = {}
largest_expense = 0
largest_category = ""

for e in expenses:
    if e["category"] not in category_totals:
        category_totals[e["category"]] = e["amount"]
    else:
        category_totals[e["category"]] += e["amount"]
    
    if e["amount"] > largest_expense:
        largest_expense = e["amount"]
        largest_category = e["category"]

total_spent = sum(category_totals.values())

print("Total per Category:", category_totals)
print("Total Spending:", total_spent)

if total_spent > budget:
    print("Budget Exceeded by", total_spent - budget)
else:
    print("Within Budget")

print("Largest Single Expense:", largest_category, "-", largest_expense)
```

# Problem 16 – Integrated Real-World System: Smart Business Manager

```python
transactions = []

def add_transaction(date, transaction_type, amount, category):
    transaction = {
        "date": date,
        "type": transaction_type,
        "amount": amount,
        "category": category
    }
    transactions.append(transaction)
    print(f"Added: {date} - {transaction_type} - {category} - Rs.{amount}")

def total_summary():
    total_income = 0
    total_expense = 0
    
    for transaction in transactions:
        if transaction["type"].lower() == "income":
            total_income += transaction["amount"]
        elif transaction["type"].lower() == "expense":
            total_expense += transaction["amount"]
    
    net_balance = total_income - total_expense
    
    print("\n" + "="*40)
    print("FINANCIAL SUMMARY")
    print("="*40)
    print(f"Total Income:    Rs.{total_income}")
    print(f"Total Expenses:  Rs.{total_expense}")
    print(f"Net Balance:     Rs.{net_balance}")
    print("="*40)

def category_summary():
    category_totals = {}
    
    for transaction in transactions:
        category = transaction["category"]
        amount = transaction["amount"]
        
        if category in category_totals:
            category_totals[category] += amount
        else:
            category_totals[category] = amount
    
    print("\nCATEGORY SUMMARY:")
    for category, total in category_totals.items():
        print(f"{category}: Rs.{total}")

def highest_expense():
    highest = None
    highest_amount = 0
    
    for transaction in transactions:
        if transaction["type"].lower() == "expense" and transaction["amount"] > highest_amount:
            highest = transaction
            highest_amount = transaction["amount"]
    
    if highest:
        print(f"\nBIGGEST EXPENSE:")
        print(f"On {highest['date']} for {highest['category']}: Rs.{highest['amount']}")
    else:
        print("\nNo expenses found")

def view_all_transactions():
    if not transactions:
        print("\nNo transactions yet")
        return
    
    print("\nALL TRANSACTIONS:")
    for transaction in transactions:
        print(f"{transaction['date']} - {transaction['type']} - {transaction['category']} - Rs.{transaction['amount']}")

def monthly_summary():
    monthly_totals = {}
    
    for transaction in transactions:
        month = transaction["date"][3:]
        
        if month not in monthly_totals:
            monthly_totals[month] = {"income": 0, "expense": 0}
        
        if transaction["type"].lower() == "income":
            monthly_totals[month]["income"] += transaction["amount"]
        else:
            monthly_totals[month]["expense"] += transaction["amount"]
    
    print("\nMONTHLY REPORT:")
    for month, totals in monthly_totals.items():
        net = totals["income"] - totals["expense"]
        print(f"{month}: Income Rs.{totals['income']}, Expense Rs.{totals['expense']}, Net Rs.{net}")

add_transaction("01-01-2024", "Income", 50000, "Salary")
add_transaction("05-01-2024", "Expense", 15000, "Rent")
add_transaction("10-01-2024", "Expense", 5000, "Food")
add_transaction("15-01-2024", "Income", 10000, "Freelance")
add_transaction("20-01-2024", "Expense", 3000, "Transport")
add_transaction("25-01-2024", "Expense", 8000, "Entertainment")
add_transaction("01-02-2024", "Income", 45000, "Salary")
add_transaction("08-02-2024", "Expense", 12000, "Shopping")

print("SMART BUSINESS MANAGER")
print("======================")

view_all_transactions()
total_summary()
category_summary()
highest_expense()
monthly_summary()

print("\nADDING MORE DATA...")
add_transaction("12-02-2024", "Expense", 2500, "Food")
add_transaction("18-02-2024", "Income", 8000, "Bonus")
add_transaction("25-02-2024", "Expense", 12000, "Electronics")

print("\nUPDATED REPORT:")
total_summary()
category_summary()
highest_expense()
```
