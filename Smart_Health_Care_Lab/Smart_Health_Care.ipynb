```python
clean_patients = []

print("--- Q1: Converting Data ---")
for row in patients_data:
    patient_dict = {}
    # Split by comma to get fields
    parts = row.split(',')
    
    for part in parts:
        # Split by colon to get key and value
        if ':' in part:
            key, value = part.split(':', 1)
            key = key.strip()
            value = value.strip()
            
            # Convert numbers to integers [cite: 22]
            if key in ['id', 'age', 'hr', 'glucose']:
                patient_dict[key] = int(value)
            else:
                patient_dict[key] = value
                
    clean_patients.append(patient_dict)

print(f"Successfully converted {len(clean_patients)} records.")
print(clean_patients[0])
```

# Q3 Average Heart Rate Per Doctor

```python
print("\n--- Advanced Grouping Analysis ---")
doc_map = {} 

for p in clean_patients:
    doc = p['doctor']
    if doc not in doc_map:
        doc_map[doc] = []
    doc_map[doc].append(p)
print("\nQ3 & Q13 Doctor Summaries:")
for doc, patients in doc_map.items():
    count = len(patients)
    avg_doc_hr = sum(p['hr'] for p in patients) / count
    avg_doc_gluc = sum(p['glucose'] for p in patients) / count
    
    print(f"Doctor: {doc}")
    print(f" - Patients: {count}")
    print(f" - Avg HR: {avg_doc_hr:.1f}")
    print(f" - Avg Glucose: {avg_doc_gluc:.1f}")
```
